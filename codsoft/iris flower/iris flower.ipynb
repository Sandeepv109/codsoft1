```python
x=data.drop(columns="species")
```

```python
y=data['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier()
```

```python
KNN.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
```

```python
LR_accuracy=LR.score(x_test,y_test)*100
KNN_accuracy=KNN.score(x_test,y_test)*100
DT_accuracy=DT.score(x_test,y_test)*100
```

```python
print(f"Accuracy by using Logistic Regression:{LR_accuracy}%")
```

```python
print(f"Accuracy by using Logistic Regression:{KNN_accuracy}%")
```

```python
print(f"Accuracy by using Logistic Regression:{DT_accuracy}%")
```
